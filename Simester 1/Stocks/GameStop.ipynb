Date - the day we observed the stock.
Open - The price of the stock the mark opens.
High - The highest recorded price of the stock that day.
Low - The lowest price of the day.
close - The price when the stock market closes.
Adj Close - The adjusted price of a stock when the market closes.
Volume - The number of the stocks being traded in  

```python
df[df["High"] == df["High"].max()]
```

```python
count = 0
total = 0
for month in range(1, 13):
    for i in range(len(df)):
        if df.loc[i]["Date"].month == month:
            avg = (df.loc[i]["High"] + df.loc[i]["Low"]) / 2
            total += avg
            count += 1
total/count
```
